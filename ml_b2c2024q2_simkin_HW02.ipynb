4. Регрессионное дерево выдает среднее от всех объектов в листе, в этом случаее квадратичное отклонение ошибок прогноза будет минимально.
5. 
#### Первое разбиение
$$
n = 10 + 30 + 20 + 40 = 100\\
L = 40 ; R = 60\\
L: p_{0l} = \frac{10}{40} = \frac{1}{4} ; p_{1l} = \frac{30}{40} = \frac{3}{4}\\
R: p_{0r} = \frac{20}{60} = \frac{1}{3} ; p_{1r} = \frac{40}{60} = \frac{2}{3}\\
H(L) = - p_{0l} \cdot ln(p_{0l}) - p_{1l} \cdot ln(p_{1l}) = 0.5623351\\
H(R) = - p_{0r} \cdot ln(p_{0r}) - p_{1r} \cdot ln(p_{1r}) = 0.6365141\\
G(t_1) = \frac{L}{n} \cdot H(L) + \frac{R}{n} \cdot H(R) = 0.606842559\\
$$
#### Второе разбиение
$$
n = 10 + 30 + 20 + 40 = 100\\
L = 30 ; R = 70\\
L: p_{0l} = \frac{20}{30} = \frac{2}{3} ; p_{1l} = \frac{10}{30} = \frac{1}{3}\\
R: p_{0r} = \frac{10}{70} = \frac{1}{7} ; p_{1r} = \frac{60}{70} = \frac{6}{7}\\
H(L) = - p_{0l} \cdot ln(p_{0l}) - p_{1l} \cdot ln(p_{1l}) = 0.6365141\\
H(R) = - p_{0r} \cdot ln(p_{0r}) - p_{1r} \cdot ln(p_{1r}) = 0.4101163\\
G(t_2) = \frac{L}{n} \cdot H(L) + \frac{R}{n} \cdot H(R) = 0.47803567\\
$$
#### Результат
$$
min(G(t_1), G(t_2)) = 0.47803567 \\
$$
<font color="Red">Ответ: 2е разбиение лучше (t2) </font>

```python
# 5 
# Первое разбиение
n = 100
p0_L=1/4
p1_L=3/4
p0_R=1/3
p1_R=2/3
L = 40
R = 60
HL = -p0_L*ln(p0_L) - p1_L*ln(p1_L)
HR = -p0_R*ln(p0_R) - p1_R*ln(p1_R)
G = HL*L/n + HR*R/n
print(f' 5. Первое разбиение H(L) = {HL}, H(R) = {HR}, G(t) = {G}')

# Второе разбиение

n = 100
p0_L=1/3
p1_L=2/3
p0_R=1/7
p1_R=6/7
L = 30
R = 70
HL = -p0_L*ln(p0_L) - p1_L*ln(p1_L)
HR = -p0_R*ln(p0_R) - p1_R*ln(p1_R)
G = HL*L/n + HR*R/n
print(f' 5. Второе разбиение H(L) = {HL}, H(R) = {HR}, G(t) = {G}')
```

```python
# 6
#Первое разбиение: [0; 30; 40; 50] и [40; 25; 45; 45]
#Второе разбиение: [20; 10; 40;10] и [20; 45; 45; 85]
# 6
list_L1 = [0, 30, 40, 50]
list_R1 = [40, 25, 45, 45]

list_L2 = [20, 10, 40, 10]
list_R2 = [20, 45, 45, 85]
#print (sum(list_L1) + sum(list_R1), sum(list_L2) + sum(list_R2))
def H(data: list) -> float :
    sum_of_values = sum(data)
    p = [x / sum_of_values for x in data]
    H = sum([-x*ln(x) if x > 0 else 0 for x in p])
    print(f'H {H}')
    return H

def G(list_L: list, list_R: list) -> float:
    HR = H(list_R)
    HL = H(list_L)
    R = sum(list_R)
    L = sum(list_L)
    print (f'L {L}, R {R}, n {L + R}')
    G = HL*(L/(L+R)) + HR*(R/(L+R))
    return G

Gt1 = G(list_L1, list_R1)
Gt2 = G(list_L2, list_R2)
print (f'G(t1) = {Gt1}, G(t2) = {Gt2}')

#1.2550396103499544
```

```python
def fill_data(data, columns_dict):
    for key, columns in columns_dict.items():
        if key == 'median':
            data[columns] = data[columns].fillna(data[columns].median())
        if key == 'freq':
            data[columns] = data[columns].fillna(data[columns].value_counts().idxmax())
            
dict_to_clean = {'median': ["Age", "Fare"], 'freq':"Embarked"}        

fill_data(test_df, dict_to_clean)
fill_data(train_df, dict_to_clean)
```

## 4. Дерево решений с настройкой параметров (30%)

Обучите на имеющейся выборке дерево решений (`DecisionTreeClassifier`). Также укажите `random_state=17`. Поэкспериментируейте с максимальной глубиной и минимальным числом элементов в листе

```python
tree.DecisionTreeClassifier?
```

```python
import pprint
best_criterion='gini' # провел серию на проверку лучшего критерия, gini во всех случаях лучший
answers = []

for test_size in [x / 10.0 for x in range(1, 5, 1)]:
    for depth in range(1,5):
        for leaves in range(1,5):
            clf = tree.DecisionTreeClassifier(random_state = 17, 
                                                  max_depth = depth, 
                                                  min_samples_leaf = leaves, 
                                                  criterion= best_criterion)
            Xtrain, Xtest, ytrain, ytest = train_test_split(train_df, 
                                                            y, 
                                                            test_size = test_size, 
                                                            random_state = 17)
            clf.fit(Xtrain, ytrain)
            prediction = clf.predict(Xtest)
            accuracy = accuracy_score(ytest, prediction)
            cross_val_scores = model_selection.cross_val_score(clf, 
                                                               Xtest, 
                                                               ytest, 
                                                               scoring="accuracy", 
                                                               cv=5)
            answers.append([test_size, depth, leaves, accuracy, cross_val_scores.mean(), cross_val_scores.std()])
            
df = pd.DataFrame(answers)
print ('Best cross_val_score')
pprint.pprint(df[df[4] == df[4].max()])
print ('Best accuracy')
pprint.pprint(df[df[3] == df[3].max()])
print ('Min std')
pprint.pprint(df[df[5] == df[5].min()])
```
